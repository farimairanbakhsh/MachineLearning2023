```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.datasets import make_classification, make_blobs, make_circles
```

```python
X, y= make_classification(n_samples=1000,
                          n_features=2,
                          n_redundant=0,
                          n_classes=2,
                          n_clusters_per_class=1,
                          class_sep=2,
                          random_state=83)

colors = np.array(['blue','red'])
plt.scatter(X[:,0],X[:,1], c=colors[y])
plt.show
```

```python
x_train,x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model = LogisticRegression(solver='sag', max_iter=200, random_state=83)
model.fit(x_train, y_train)
model.predict(x_test), y_test
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model1 = SGDClassifier(loss= 'log_loss', random_state=83)
model1.fit(x_train, y_train)
```

```python
model1.score(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
x1_min, x2_min = X.min(0)
x1_max, x2_max = X.max(0)

n=500
x1r = np.linspace(x1_min, x1_max, n)
x2r = np.linspace(x2_min, x2_max, n)
x1m , x2m = np.meshgrid(x1r, x2r)

xm = np.stack((x1m.flatten(), x2m.flatten()), axis=1)
ym = model.decision_function(xm)

predictions = model.predict(X)
misclassified_indices = np.where(predictions != y)[0]
colors = np.array(['blue', 'red'])

plt.scatter(X[predictions == y, 0], X[predictions == y, 1], c=colors[y[predictions == y]])
plt.scatter(X[misclassified_indices, 0], X[misclassified_indices, 1], marker='x', c='green', label='Misclassified')

plt.contour(x1m, x2m, ym.reshape(x1m.shape), levels=[-1, 0, 1])

plt.legend()
plt.show()
```

```python
X, y= make_classification(n_samples=1000,
                          n_features=2,
                          n_redundant=0,
                          n_classes=3,
                          n_clusters_per_class=1,
                          class_sep=2,
                          random_state=83)

colors = np.array(['blue','red', 'orange'])
plt.scatter(X[:,0],X[:,1], c=colors[y])
plt.show
```

```python
x_train,x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(solver='sag', max_iter=200, random_state=27)
model.fit(x_train, y_train)
model.predict(x_test), y_test
```

```python
x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max, 500),
                       np.linspace(x2_min, x2_max, 500))

Z = model.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z = Z.reshape(xx1.shape)

plt.contourf(xx1, xx2, Z, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')

predicted_classes = model.predict(X)
misclassified_indices = np.where(predicted_classes != y)[0]
misclassified_points = X[misclassified_indices]

plt.scatter(misclassified_points[:, 0], misclassified_points[:, 1], marker='x', c='red', label='Misclassified')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundaries with Misclassified Points')
plt.legend()
plt.show()
```
